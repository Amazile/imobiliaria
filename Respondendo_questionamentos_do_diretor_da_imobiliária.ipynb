```python
#Quantos imóveis apresentam no mínimo 4 quartos e aluguel menor que R$2.000,00?
df_especial=(df['Quartos']>= 4) & (df['Valor']<2000.0)
df_especial_quartos_valor=df[df_especial].shape[0]
df_especial_quartos_valor
```

```python
#Criar coluna com o valor bruto do aluguel (valor+condomínio+IPTU)
df['Valor bruto aluguel']=df['Valor']+df['Condominio']+df['IPTU']
df.head(10)
```

```python
#Criar coluna com valor/m2
df['Valor m2']=df['Valor']/df['Area']
df.head(10)
```

```python
#arredondando valor m2 para duas casas decimais
df['Valor m2']=df['Valor m2'].round(2)
df.head()
```

```python
#selecionando casa, casa de condomínio e casa de vila
residencial_casa=['Casa','Casa de Condomínio', 'Casa de Vila']
selecao_casa=df['Tipo'].isin(residencial_casa)
dados_residencial_casa=df[selecao_casa]
```

```python
dados_residencial_casa.shape
print('A base de dados apresenta {} registros de casa/casa de condomínio/casa de vila.'.format(dados_residencial_casa.shape[0]))
```
